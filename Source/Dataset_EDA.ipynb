```python
# -----------------------------
# Drive 마운트 + .gz 압축해제 유틸
# -----------------------------

def mount_drive_if_colab(mount_point: str = "/content/drive") -> None:
    if "google.colab" in sys.modules:
        from google.colab import drive
        drive.mount(mount_point)
    else:
        print("로컬 환경이라 drive.mount 생략")

def extract_gz_folder(source_folder: str, dest_folder: str, overwrite: bool = False) -> List[str]:
    """
    source_folder 내 *.gz 를 dest_folder에 압축해제
    반환: 생성된 파일 경로 리스트
    """
    source_folder = str(source_folder)
    dest_folder = str(dest_folder)
    os.makedirs(dest_folder, exist_ok=True)

    files = os.listdir(source_folder)
    gz_files = [f for f in files if f.endswith(".gz")]

    if not gz_files:
        print(f"경고: {source_folder} 에 .gz 파일이 없음")
        return []

    created = []
    print(f"[extract] {source_folder} -> {dest_folder}  (gz files={len(gz_files)})")

    for file_name in gz_files:
        src = os.path.join(source_folder, file_name)
        out_name = file_name[:-3]  # remove .gz
        dst = os.path.join(dest_folder, out_name)

        if (not overwrite) and os.path.exists(dst) and os.path.getsize(dst) > 0:
            # 이미 풀려있으면 스킵
            continue

        try:
            with gzip.open(src, "rb") as f_in, open(dst, "wb") as f_out:
                shutil.copyfileobj(f_in, f_out)
            created.append(dst)
        except Exception as e:
            print(f"에러: {file_name} -> {e}")

    print(f"완료: 새로 생성/갱신 {len(created)}개")
    return created

# 실행
mount_drive_if_colab()
_ = extract_gz_folder(DRIVE_SOURCE_FOLDER, LOCAL_DATA_DIR, overwrite=False)
```

# 코호트(STAY) 생성
  - ICU내 사망 + 성인/60일 제한 + Full code인 환자

```python
# 1) 기본 코호트(성인 + ICU 60일 이하 + ICU 내 사망 플래그)
con.execute("""
CREATE OR REPLACE TABLE cohort_stays_base AS
SELECT
  ie.subject_id,
  ie.hadm_id,
  ie.stay_id,
  ie.intime::TIMESTAMP AS intime,
  ie.outtime::TIMESTAMP AS outtime,
  a.deathtime::TIMESTAMP AS deathtime,
  p.gender AS sex,
  p.anchor_age::INTEGER AS age,
  CASE
    WHEN a.deathtime IS NOT NULL
     AND a.deathtime::TIMESTAMP BETWEEN ie.intime::TIMESTAMP AND ie.outtime::TIMESTAMP
    THEN 1 ELSE 0 END AS death_in_icu
FROM icustays ie
JOIN admissions a ON ie.hadm_id = a.hadm_id
JOIN patients p ON ie.subject_id = p.subject_id
WHERE p.anchor_age::INTEGER >= 18
  AND (epoch(ie.outtime::TIMESTAMP - ie.intime::TIMESTAMP)/86400.0) <= 60
;
""")

# 2) "Full code" stay 식별 (itemid=223758 AND value='Full code')
#    - ICU stay 시간 범위 내 기록만 인정
#    - value 대소문자/공백 차이 대응
con.execute("""
CREATE OR REPLACE TABLE fullcode_stays AS
SELECT DISTINCT cs.stay_id
FROM cohort_stays_base cs
JOIN chartevents ce
  ON ce.stay_id = cs.stay_id
WHERE ce.itemid = 223758
  AND ce.value IS NOT NULL
  AND ce.charttime::TIMESTAMP BETWEEN cs.intime AND cs.outtime
  AND lower(trim(ce.value)) = 'full code'
;
""")

# 3) 최종 코호트 = Full code stay만 남김
con.execute("""
CREATE OR REPLACE TABLE cohort_stays AS
SELECT *
FROM cohort_stays_base
WHERE stay_id IN (SELECT stay_id FROM fullcode_stays)
;
""")

# 4) 확인
con.execute("""
SELECT
  (SELECT COUNT(*) FROM cohort_stays_base) AS n_base,
  (SELECT COUNT(*) FROM fullcode_stays) AS n_fullcode_stays,
  (SELECT COUNT(*) FROM cohort_stays) AS n_final,
  (SELECT SUM(death_in_icu) FROM cohort_stays) AS deaths_in_icu_final
;
""").fetchdf()
```

## 코호트 기본 규모 통계 시각화

```python
# 0) cohort 규모 통계 먼저 계산해서 파이썬 변수로 꺼내기
stats = con.execute("""
SELECT
  (SELECT COUNT(*) FROM cohort_stays_base) AS n_base,
  (SELECT COUNT(*) FROM fullcode_stays) AS n_fullcode_stays,
  (SELECT COUNT(*) FROM cohort_stays) AS n_final,
  COALESCE((SELECT SUM(death_in_icu) FROM cohort_stays), 0) AS deaths_in_icu_final
;
""").fetchdf()

# 파이썬 스칼라로 언팩
n_base = int(stats.loc[0, "n_base"])
n_fullcode_stays = int(stats.loc[0, "n_fullcode_stays"])
n_final = int(stats.loc[0, "n_final"])
deaths = int(stats.loc[0, "deaths_in_icu_final"])

# 사망률 (%)
icu_mortality_rate = (deaths / n_final * 100) if n_final > 0 else 0.0

stats
```

### 지정한 itemid별 데이터 개수 출력

```python
import pandas as pd

# itemid 리스트 테이블 등록(기존 방식 유지)
con.register("vital_itemid_df", pd.DataFrame({"itemid": vital_itemid_extract}))
con.register("lab_itemid_df",   pd.DataFrame({"itemid": lab_itemid}))

con.execute("DROP TABLE IF EXISTS vital_itemids;")
con.execute("DROP TABLE IF EXISTS lab_itemids;")
con.execute("CREATE TABLE vital_itemids AS SELECT DISTINCT itemid::INTEGER AS itemid FROM vital_itemid_df;")
con.execute("CREATE TABLE lab_itemids   AS SELECT DISTINCT itemid::INTEGER AS itemid FROM lab_itemid_df;")

con.execute("DROP TABLE IF EXISTS events_long;")
con.execute("""
CREATE TABLE events_long AS
WITH
ce AS (
  SELECT
    cs.subject_id, cs.hadm_id, cs.stay_id,
    cs.intime, cs.outtime, cs.deathtime, cs.death_in_icu,
    'chartevents' AS source,
    ce.charttime::TIMESTAMP AS charttime,
    ce.storetime::TIMESTAMP AS storetime,

    -- ✅ (1) itemid 표준화: 화씨(223761)도 섭씨 체온 itemid(223762)로 합침
    CASE
      WHEN ce.itemid::INTEGER = 223761 THEN 223762
      ELSE ce.itemid::INTEGER
    END AS itemid,

    -- ✅ (2) 값 표준화: 화씨면 섭씨로 변환
    CASE
      WHEN ce.itemid::INTEGER = 223761 THEN (ce.valuenum::DOUBLE - 32.0) * (5.0/9.0)
      ELSE ce.valuenum::DOUBLE
    END AS value_conv

  FROM chartevents ce
  JOIN cohort_stays cs ON cs.stay_id = ce.stay_id
  JOIN vital_itemids vi ON vi.itemid = ce.itemid
  WHERE ce.charttime IS NOT NULL
    AND ce.valuenum IS NOT NULL
    -- ✅ (옵션이지만 강추) ICU 시간 범위로 자르기: 불필요한 row 폭증 방지
    AND ce.charttime::TIMESTAMP BETWEEN cs.intime AND cs.outtime
),

ce_filtered AS (
  SELECT
    ce.*,
    CASE
      WHEN vr.itemid IS NULL THEN ce.value_conv
      WHEN ce.value_conv BETWEEN vr.vmin AND vr.vmax THEN ce.value_conv
      ELSE NULL
    END AS value
  FROM ce
  LEFT JOIN valid_ranges vr ON vr.itemid = ce.itemid   -- ✅ 표준화된 itemid로 조인
  WHERE (vr.itemid IS NULL OR ce.value_conv BETWEEN vr.vmin AND vr.vmax)
),

le AS (
  SELECT
    cs.subject_id, cs.hadm_id, cs.stay_id,
    cs.intime, cs.outtime, cs.deathtime, cs.death_in_icu,
    'labevents' AS source,
    COALESCE(le.charttime::TIMESTAMP, le.storetime::TIMESTAMP) AS charttime,
    le.storetime::TIMESTAMP AS storetime,
    le.itemid::INTEGER AS itemid,
    le.valuenum::DOUBLE AS value_raw
  FROM labevents le
  JOIN cohort_stays cs
    ON cs.hadm_id = le.hadm_id
   AND COALESCE(le.charttime::TIMESTAMP, le.storetime::TIMESTAMP) BETWEEN cs.intime AND cs.outtime
  JOIN lab_itemids li ON li.itemid = le.itemid
  WHERE COALESCE(le.charttime::TIMESTAMP, le.storetime::TIMESTAMP) IS NOT NULL
    AND le.valuenum IS NOT NULL
),

le_filtered AS (
  SELECT
    le.*,
    CASE
      WHEN vr.itemid IS NULL THEN le.value_raw
      WHEN le.value_raw BETWEEN vr.vmin AND vr.vmax THEN le.value_raw
      ELSE NULL
    END AS value
  FROM le
  LEFT JOIN valid_ranges vr ON vr.itemid = le.itemid
  WHERE (vr.itemid IS NULL OR le.value_raw BETWEEN vr.vmin AND vr.vmax)
),

u AS (
  SELECT
    subject_id, hadm_id, stay_id,
    intime, outtime, deathtime, death_in_icu,
    source, charttime, storetime, itemid, value
  FROM ce_filtered
  UNION ALL
  SELECT
    subject_id, hadm_id, stay_id,
    intime, outtime, deathtime, death_in_icu,
    source, charttime, storetime, itemid, value
  FROM le_filtered
),

dedup AS (
  SELECT
    subject_id, hadm_id, stay_id,
    intime, outtime, deathtime, death_in_icu,
    source, charttime, itemid,
    arg_max(value, COALESCE(storetime, charttime)) AS value
  FROM u
  GROUP BY
    subject_id, hadm_id, stay_id,
    intime, outtime, deathtime, death_in_icu,
    source, charttime, itemid
)

SELECT * FROM dedup;
""")

con.execute("SELECT source, COUNT(*) AS n FROM events_long GROUP BY source;").fetchdf()
```

# 이벤트 기반 time_index + delta 파생변수 생성

```python
# time_index
con.execute("DROP TABLE IF EXISTS time_index;")
con.execute("""
CREATE TABLE time_index AS
SELECT stay_id, charttime
FROM events_long
GROUP BY stay_id, charttime;
""")

# event_flags
con.execute("DROP TABLE IF EXISTS event_flags;")
con.execute("""
CREATE TABLE event_flags AS
SELECT
  stay_id,
  charttime,
  MAX(CASE WHEN source='chartevents' THEN 1 ELSE 0 END) AS vital_flag,
  MAX(CASE WHEN source='labevents'  THEN 1 ELSE 0 END) AS lab_flag
FROM events_long
GROUP BY stay_id, charttime;
""")

# deltas
con.execute("DROP TABLE IF EXISTS deltas;")
con.execute("""
CREATE TABLE deltas AS
SELECT
  stay_id,
  charttime,
  date_diff(
    'second',
    MAX(CASE WHEN vital_flag=1 THEN charttime END)
      OVER (PARTITION BY stay_id ORDER BY charttime ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW),
    charttime
  )/3600.0 AS delta_vital_hr,

  date_diff(
    'second',
    MAX(CASE WHEN lab_flag=1 THEN charttime END)
      OVER (PARTITION BY stay_id ORDER BY charttime ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW),
    charttime
  )/3600.0 AS delta_lab_hr
FROM event_flags;
""")
```

# wide pivot
  - 행으로 들어가있는 itemid를 컬럼으로 변경하는 과정

```python
# pivot 컬럼 SQL 생성
all_itemids = sorted(set(vital_itemid_feature) | set(lab_itemid))

pivot_cols_sql = ",\n  ".join([
    f"MAX(CASE WHEN e.itemid = {iid} THEN e.value END) AS item_{iid}"
    for iid in all_itemids
])

con.execute("DROP TABLE IF EXISTS timeseries_wide;")
con.execute(f"""
CREATE TABLE timeseries_wide AS
SELECT
  cs.subject_id, cs.hadm_id, cs.stay_id,
  ti.charttime,
  cs.intime, cs.outtime, cs.deathtime, cs.death_in_icu,
  cs.age, cs.sex,
  {pivot_cols_sql}
FROM time_index ti
JOIN cohort_stays cs ON cs.stay_id = ti.stay_id
LEFT JOIN events_long e ON e.stay_id=ti.stay_id AND e.charttime=ti.charttime
GROUP BY
  cs.subject_id, cs.hadm_id, cs.stay_id,
  ti.charttime,
  cs.intime, cs.outtime, cs.deathtime, cs.death_in_icu,
  cs.age, cs.sex
;
""")

con.execute("DROP TABLE IF EXISTS dataset_base;")
con.execute("""
CREATE TABLE dataset_base AS
SELECT
  t.*,
  d.delta_vital_hr,
  d.delta_lab_hr
FROM timeseries_wide t
LEFT JOIN deltas d
  ON d.stay_id=t.stay_id AND d.charttime=t.charttime
;
""")
```

# 논문 방식 트리밍 + 라벨(y)
  - 비식별화된 시간 데이터를 실제 데이터로 변경 및 타임스탬프 별 사망 여부 피쳐 생성

```python
con.execute("DROP TABLE IF EXISTS dataset_final;")
con.execute("""
CREATE TABLE dataset_final AS
SELECT
  *,
  CASE
    WHEN death_in_icu=1
     AND charttime BETWEEN (deathtime - INTERVAL '24 hours') AND deathtime
    THEN 1 ELSE 0 END AS y
FROM dataset_base
WHERE
  (death_in_icu=0)
  OR (death_in_icu=1 AND charttime BETWEEN (deathtime - INTERVAL '24 hours') AND deathtime)
;
""")

con.execute("SELECT COUNT(*) AS n_rows, SUM(y) AS n_event_rows FROM dataset_final;").fetchdf()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------
# 0) 스타일 (폰트 포함)
# -----------------------------
sns.set_theme(style="white", context="talk")
setup_korean_font_colab()
plt.rcParams["axes.unicode_minus"] = False

# -----------------------------
# 1) itemid 목록 준비 (+ 체온 화씨->섭씨 통일)
#    - 추출에는 223761(화씨) 포함되어도 됨
#    - 피쳐에서는 223761 제거하고 223762(섭씨)로 통일
# -----------------------------
vital_itemid_extract = list(vital_itemid_extract)  # 이미 있음
vital_itemid_feature = list(vital_itemid_feature)  # 이미 있음
lab_itemid = list(lab_itemid)                      # 이미 있음

# (중요) 섭씨/화씨 통일: 피쳐는 무조건 223762로만 가게 정규화
def normalize_temp_itemids(ids):
    ids = [223762 if int(x) == 223761 else int(x) for x in ids]
    # 중복 제거
    return sorted(set(ids))

vital_itemid_feature_norm = normalize_temp_itemids(vital_itemid_feature)
vital_itemid_extract_norm = sorted(set([int(x) for x in vital_itemid_extract]))  # 추출은 그대로 둬도 됨
lab_itemid_norm = sorted(set([int(x) for x in lab_itemid]))

ALL_ITEMIDS = sorted(set(vital_itemid_feature_norm) | set(lab_itemid_norm))

# -----------------------------
# 2) d_items / d_labitems에서 라벨/단위 자동 매칭
# -----------------------------
# vital(=chartevents) 라벨/단위
if vital_itemid_extract_norm:
    vital_meta = con.execute(f"""
        SELECT
            itemid::INTEGER AS itemid,
            label,
            unitname,
            'vital(chartevents)' AS source
        FROM d_items
        WHERE itemid IN ({",".join(map(str, vital_itemid_extract_norm))})
        ORDER BY itemid
    """).fetchdf()
else:
    vital_meta = pd.DataFrame(columns=["itemid","label","unitname","source"])

# lab 라벨/단위
if lab_itemid_norm:
    lab_meta = con.execute(f"""
        SELECT
            itemid::INTEGER AS itemid,
            label,
            NULL AS unitname,
            'lab(labevents)' AS source
        FROM d_labitems
        WHERE itemid IN ({",".join(map(str, lab_itemid_norm))})
        ORDER BY itemid
    """).fetchdf()
else:
    lab_meta = pd.DataFrame(columns=["itemid","label","unitname","source"])

meta = pd.concat([vital_meta, lab_meta], ignore_index=True)

# (선택) 체온 라벨/단위가 이상하면 강제로 정리
# 223762가 meta에 없거나 label이 비어있을 때도 대비
if 223762 in ALL_ITEMIDS:
    if meta.loc[meta["itemid"] == 223762].empty:
        meta = pd.concat([meta, pd.DataFrame([{
            "itemid": 223762,
            "label": "Temperature (C)",
            "unitname": "°C",
            "source": "vital(chartevents)"
        }])], ignore_index=True)
    else:
        # label/unitname 비어있으면 보정
        idx = meta.index[meta["itemid"] == 223762]
        meta.loc[idx, "label"] = meta.loc[idx, "label"].fillna("Temperature (C)")
        meta.loc[idx, "unitname"] = meta.loc[idx, "unitname"].fillna("°C")

# -----------------------------
# 3) itemid -> pivot 컬럼명(item_XXXX) & 사람이 읽는 feature명 생성
# -----------------------------
meta["col"] = meta["itemid"].apply(lambda x: f"item_{int(x)}")

def make_feature_name(row):
    # label이 있으면 label 쓰고, unit이 있으면 붙임
    label = str(row["label"]) if pd.notna(row["label"]) else f"itemid={row['itemid']}"
    unit = str(row["unitname"]) if pd.notna(row["unitname"]) and str(row["unitname"]).strip() != "" else None
    if unit:
        return f"{label} ({unit})"
    return label

meta["feature"] = meta.apply(make_feature_name, axis=1)

# -----------------------------
# 4) MAIN_FEATURES 자동 생성
#    - age는 itemid가 아니라 cohort 컬럼이라 수동 추가
#    - 나머지는 ALL_ITEMIDS에서 생성
# -----------------------------
MAIN_FEATURES = {"Age": "age"}
for _, r in meta.drop_duplicates("col").iterrows():
    # 예: "Heart Rate (/min)" 같은게 key로 들어감
    MAIN_FEATURES[r["feature"]] = r["col"]

# 매핑 테이블 확인용
meta_display = meta[["source","itemid","col","feature"]].drop_duplicates().sort_values(["source","itemid"]).reset_index(drop=True)
meta_display
```

```python
# -----------------------------
# 5) dataset_final 스키마 검사
# -----------------------------
schema = con.execute("SELECT name FROM pragma_table_info('dataset_final')").fetchdf()["name"].tolist()
missing_cols = [v for v in MAIN_FEATURES.values() if v not in schema]

if missing_cols:
    raise KeyError(f"dataset_final에 없는 컬럼이 있음: {missing_cols}\n"
                   f"원인: pivot 단계 itemid 목록/정규화(223761->223762) 불일치 가능")

# -----------------------------
# 6) 결측률 계산 (DuckDB)
# -----------------------------
feature_cols = [c for c in MAIN_FEATURES.values() if c != "age"]
exprs = ",\n".join([f"AVG(CASE WHEN {col} IS NULL THEN 1 ELSE 0 END) AS {col}" for col in feature_cols] + [
    "AVG(CASE WHEN age IS NULL THEN 1 ELSE 0 END) AS age"
])

miss = con.execute(f"SELECT {exprs} FROM dataset_final;").fetchdf().T.reset_index()
miss.columns = ["col", "missing_rate"]
miss["missing_pct"] = miss["missing_rate"] * 100

# 사람이 읽는 이름 역매핑
rev_map = {v: k for k, v in MAIN_FEATURES.items()}
miss["feature"] = miss["col"].map(rev_map).fillna(miss["col"])
miss = miss.sort_values("missing_pct", ascending=False).reset_index(drop=True)

# -----------------------------
# 7) 시각화 A: 주요 피처 결측률
# -----------------------------
fig, ax = plt.subplots(figsize=(12, 8))
sns.barplot(
    data=miss, x="missing_pct", y="feature",
    palette="flare", edgecolor="#333333", linewidth=1, ax=ax
)
ax.grid(axis='x', linestyle='--', alpha=0.5)
sns.despine(left=True)
ax.set_title("피처별 결측률 (Missingness)", pad=25, fontsize=20, fontweight='bold')
ax.set_xlabel("Missing Rate (%)", fontsize=14)
ax.set_ylabel("")

for i, r in miss.iterrows():
    ax.text(r["missing_pct"] + 0.8, i, f'{r["missing_pct"]:.1f}%',
            va="center", fontsize=12, fontweight='bold', color='#C0392B')

plt.tight_layout()
plt.show()

# -----------------------------
# 8) 분포 확인용 샘플링 (DuckDB)
# -----------------------------
cols_sql = ", ".join(MAIN_FEATURES.values())
sample_df = con.execute(f"""
    SELECT {cols_sql}
    FROM dataset_final
    USING SAMPLE 30 PERCENT (bernoulli);
""").fetchdf()

long_df = sample_df.melt(var_name="col", value_name="value")
long_df["feature"] = long_df["col"].map(rev_map).fillna(long_df["col"])

# delta 컬럼은 네 dataset_final 컬럼명 기준으로 판별
# (너는 delta_vital_hr, delta_lab_hr 로 만들었음)
long_df["is_delta"] = long_df["col"].str.contains("delta_", case=False, na=False)

# -----------------------------
# 9) 시각화 B-1: 일반 변수 분포 (KDE FacetGrid)
# -----------------------------
normal_features = long_df[~long_df["is_delta"]].dropna()

g = sns.FacetGrid(
    normal_features, col="feature", col_wrap=4,
    sharex=False, sharey=False, height=3, aspect=1.3
)
g.map(sns.kdeplot, "value", fill=True, alpha=0.5, linewidth=2)
g.set_titles("{col_name}", fontweight='bold', size=12)
g.fig.suptitle("변수별 관측치 분포 (KDE)", y=1.05, fontsize=22, fontweight='bold')

for ax in g.axes.flat:
    ax.grid(axis='both', linestyle=':', alpha=0.35)
    ax.tick_params(labelsize=9)

plt.tight_layout()
plt.show()

# -----------------------------
# 10) 시각화 B-2: Delta 분포 (로그 스케일)
# -----------------------------
delta_df = long_df[long_df["is_delta"]].dropna()

if len(delta_df) > 0:
    fig, ax = plt.subplots(figsize=(12, 6))
    sns.violinplot(
        data=delta_df, x="feature", y="value",
        palette="Pastel1", inner="quartile", ax=ax
    )
    ax.set_yscale("log")
    ax.set_title("관측 간격(Delta) 분포 (Log Scale)", pad=25, fontsize=20, fontweight='bold')
    ax.set_ylabel("Hours (Log Scale)", fontsize=14)
    ax.grid(axis='y', which='both', linestyle='--', alpha=0.3)
    sns.despine()
    plt.tight_layout()
    plt.show()
else:
    print("delta_* 컬럼이 없거나 샘플에서 전부 결측이라 Delta 시각화를 건너뜀")

# -----------------------------
# 11) 시각화 C: Stay 단위 데이터 밀도
# -----------------------------
stay_obs = con.execute("""
    SELECT stay_id, COUNT(*) AS n_rows
    FROM dataset_final
    GROUP BY stay_id
""").fetchdf()

fig, ax = plt.subplots(figsize=(12, 6))
sns.histplot(stay_obs["n_rows"], bins=50, kde=True, edgecolor="white", alpha=0.5, ax=ax)
ax.set_xscale("log")
ax.set_title("ICU stay별 관측 시점 수 분포", pad=25, fontsize=20, fontweight='bold')
ax.set_xlabel("Rows per stay (log scale)", fontsize=14)
ax.set_ylabel("Number of stays", fontsize=14)

med = float(stay_obs["n_rows"].median())
ax.axvline(med, color='#333333', linestyle='--', linewidth=2)
ax.annotate(f'Median: {med:.0f} points',
            xy=(med, ax.get_ylim()[1]*0.75),
            xytext=(med*1.5, ax.get_ylim()[1]*0.85),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=6),
            fontsize=12, fontweight='bold')

ax.grid(axis='y', linestyle='--', alpha=0.3)
sns.despine()
plt.tight_layout()
plt.show()
```

## 시간 구조 관련 통계
  - 환자 1명당 평균 관측 시점 수: 그 환자의 stay별 관측 시점 수(n_timepoints) 를 구한 다음 평균(mean)

- 환자 1명당 median observation window 길이(hours): 그 환자의 stay별 window 길이(window_hours) 를 구한 다음 중앙값(median)

  - window는 각 stay에서 max(charttime) - min(charttime) 기준

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------
# 0) 스타일 및 테마 설정 (테마 -> 폰트 순서 엄수)
# -----------------------------
sns.set_theme(style="white", context="talk")
setup_korean_font_colab() # 사용자 정의 함수
plt.rcParams["axes.unicode_minus"] = False

# 시각화 공통 컬러 설정
COLOR_STAY = "#3498DB"   # Stay 관련 (블루)
COLOR_SUBJ = "#1ABC9C"   # Patient 관련 (민트)
COLOR_OUTCOME = {"생존": "#79B4B7", "사망": "#E98580"}
COLOR_MAIN = "#2C3E50"

# -----------------------------
# 1) 데이터 요약 표 (Patient-level & Stay-level)
# -----------------------------
# 가독성을 높이기 위해 소수점 정리 및 콤마 추가
summary_p_df = pd.DataFrame([summary_patient]).T.rename(columns={0: "값"})
summary_s_df = pd.DataFrame([summary_stay]).T.rename(columns={0: "값"})

print("### [환자 단위 요약 통계] ###")
display(summary_p_df.style.format("{:,.2f}"))
print("\n### [Stay 단위 요약 통계] ###")
display(summary_s_df.style.format("{:,.2f}"))

# -----------------------------
# 2) 시각화 1: 환자 1명당 평균 관측 시점 수 분포
# -----------------------------
fig, ax = plt.subplots(figsize=(12, 6))

# 이상치가 너무 많을 경우를 대비해 상위 5% 커팅 (선택 사항)
upper_limit = subj_df["mean_timepoints_per_stay"].quantile(0.95)
plot_data = subj_df[subj_df["mean_timepoints_per_stay"] <= upper_limit]

sns.histplot(plot_data["mean_timepoints_per_stay"], bins=40, kde=True,
             color=COLOR_SUBJ, edgecolor="white", alpha=0.6, ax=ax)

# 중앙값 가이드라인
median_val = subj_df["mean_timepoints_per_stay"].median()
ax.axvline(median_val, color=COLOR_MAIN, linestyle="--", linewidth=2)
ax.text(median_val * 1.1, ax.get_ylim()[1] * 0.8, f'중앙값: {median_val:.1f}개',
        fontweight='bold', color=COLOR_MAIN)

ax.set_title("환자별 평균 관측 시점 수 분포", pad=25, fontsize=22, fontweight="bold")
ax.set_xlabel("평균 관측 시점 수 (Timepoints per Stay)", fontsize=15)
ax.set_ylabel("환자 수 (Count)", fontsize=15)
sns.despine()
plt.tight_layout()
plt.show()

# -----------------------------
# 3) 시각화 2: 환자별 Median Observation Window(시간) 분포
# -----------------------------
fig, ax = plt.subplots(figsize=(12, 6))

# 로그 스케일 대신 현실적인 범위(예: 0~168시간=1주일)로 제한하여 가독성 증대
plot_data_window = subj_df[subj_df["median_window_hours"] <= 168]

sns.histplot(plot_data_window["median_window_hours"], bins=40, kde=True,
             color="#5DADE2", edgecolor="white", alpha=0.6, ax=ax)

# 가이드라인 (24시간, 72시간 단위)
for h in [24, 72, 120]:
    ax.axvline(h, color='gray', linestyle=':', alpha=0.5)
    ax.text(h, ax.get_ylim()[1] * 0.95, f'{h}h', ha='center', fontsize=12, color='gray')

ax.set_title("환자별 중앙 관측 기간(Window) 분포", pad=25, fontsize=22, fontweight="bold")
ax.set_xlabel("관측 기간 (Hours, 0-168h 범위)", fontsize=15)
ax.set_ylabel("환자 수 (Count)", fontsize=15)
sns.despine()
plt.tight_layout()
plt.show()

# -----------------------------
# 4) 시각화 3: Stay Window vs Timepoints (데이터 밀도 확인)
# -----------------------------
fig, ax = plt.subplots(figsize=(12, 7))

# 산점도가 너무 겹칠 때는 hexbin이나 투명도 조절이 필수
sns.scatterplot(data=stay_df, x="window_hours", y="n_timepoints",
                s=30, alpha=0.2, color=COLOR_STAY, edgecolor=None, ax=ax)

# [가시성 개선] 로그 스케일 유지하되, 주요 지점에 레이블 추가
ax.set_xscale("log")
ax.set_yscale("log")

# 기준선 (예: 1시간에 1번 측정되는 라인 y=x)
line_range = np.linspace(1, stay_df["window_hours"].max(), 100)
ax.plot(line_range, line_range, color='red', linestyle='--', alpha=0.4, label="1 Timepoint / Hour")

ax.set_title("Stay 단위: 관측 기간 대비 데이터 밀도", pad=25, fontsize=22, fontweight="bold")
ax.set_xlabel("관측 기간 (Window Hours, Log Scale)", fontsize=15)
ax.set_ylabel("관측 시점 수 (Timepoints, Log Scale)", fontsize=15)
ax.legend(fontsize=12)
ax.grid(True, which="both", linestyle=":", alpha=0.2)
sns.despine()
plt.tight_layout()
plt.show()

# -----------------------------
# 5) 시각화 4: Outcome별 Observation Window 비교 (방어용)
# -----------------------------
if has_death_in_icu:
    fig, ax = plt.subplots(figsize=(10, 7))

    tmp = stay_df.copy()
    tmp["outcome"] = np.where(tmp["death_in_icu"] == 1, "사망", "생존")

    # Boxenplot은 분포의 꼬리를 더 잘 보여줌 (로그 스케일 제거 후 120시간 제한)
    tmp_filtered = tmp[tmp["window_hours"] <= 120]

    sns.boxenplot(data=tmp_filtered, x="outcome", y="window_hours",
                  palette=COLOR_OUTCOME, width=0.5, ax=ax)

    ax.set_title("결과(Outcome)별 관측 기간 비교", pad=25, fontsize=22, fontweight="bold")
    ax.set_xlabel("ICU 결과", fontsize=15)
    ax.set_ylabel("관측 기간 (Hours)", fontsize=15)

    # 평균값 점 추가
    sns.pointplot(data=tmp_filtered, x="outcome", y="window_hours",
                  estimator=np.mean, color="white", markers="D", scale=0.7, ax=ax)

    sns.despine()
    plt.tight_layout()
    plt.show()

# -----------------------------
# 6) 시각화 5: Delta(측정 간격) 통계 (핵심!)
# -----------------------------
for dc in delta_cols:
    fig, ax = plt.subplots(figsize=(12, 6))

    # 로그 스케일 대신 현실적인 측정 간격(0~48시간) 표시
    dc_data = stay_df[dc].dropna()
    dc_filtered = dc_data[dc_data <= 48]

    sns.histplot(dc_filtered, bins=48, kde=True, color="#9B59B6", edgecolor="white", alpha=0.6, ax=ax)

    # 가이드라인 (1시간, 12시간, 24시간)
    for h in [1, 12, 24]:
        ax.axvline(h, color='red', linestyle='--', alpha=0.4)
        ax.text(h + 0.5, ax.get_ylim()[1] * 0.9, f'{h}h', color='red', fontsize=12, fontweight='bold')

    ax.set_title(f"측정 주기 분포: {dc}", pad=25, fontsize=22, fontweight="bold")
    ax.set_xlabel("측정 간격 (Hours)", fontsize=15)
    ax.set_ylabel("Stays (Count)", fontsize=15)
    sns.despine()
    plt.tight_layout()
    plt.show()
```
